```python
from pymysql import connect
from pymysql.cursors import DictCursor

class ConnectionDB:
    def __init__(self):
        self.connection = connect(
            host='localhost',
            user='root',
            password='anh1710gdt',
            db='qlsv',
            charset='utf8mb4',
            cursorclass=DictCursor)
        
    def close(self):
        self.connection.close()

    def save(self):
        self.connection.commit()

    def rollback(self):
        self.connection.rollback()

    def get_cursor(self):
        return self.connection.cursor()
    
    def get_connection(self):
        return self.connection
    
    def create_table(self, table_name, fields):
        cursor = self.get_cursor()
        cursor.execute(f'''
            CREATE TABLE IF NOT EXISTS {table_name} (
                {fields}
            )
            ''')
        self.save()
```

```python
db = ConnectionDB()
cursor = db.get_cursor()
db.create_table('khoa', '''
    id int primary key auto_increment,
    ma_khoa varchar(255) not null unique,
    name varchar(255) not null,
    created_at timestamp default current_timestamp
''')

db.create_table('nganh', '''
    id int primary key auto_increment,
    ma_nganh varchar(255) not null unique,
    name varchar(255) not null,
    created_at timestamp default current_timestamp,
    ma_khoa varchar(255) not null references khoa(ma_khoa)
''')

db.create_table('lop', '''
    id int primary key auto_increment,
    ma_lop varchar(255) not null unique,
    name varchar(255) not null,
    khoahoc varchar(50) not null,
    hedaotao varchar(50) not null,
    namhoc int not null,
    created_at timestamp default current_timestamp,
    ma_nganh varchar(255) not null references nganh(ma_nganh)
''')

db.create_table('sinhvien', '''
    id int primary key auto_increment,
    ma_sinhvien varchar(255) not null unique,
    name varchar(255) not null,
    gender boolean not null,
    brithday date not null,
    address varchar(255) not null,
    ma_lop varchar(255) not null references lop(ma_lop),
    created_at timestamp default current_timestamp
''')

db.create_table('hocphan', '''
    id int primary key auto_increment,
    ma_hocphan varchar(255) not null unique,
    name varchar(255) not null,
    sotinchi int not null,
    ma_nganh varchar(255) not null references nganh(ma_nganh),
    hocky int not null,
    created_at timestamp default current_timestamp
''')

db.create_table('diem', '''
    id int primary key auto_increment,
    diem float not null,
    ma_sinhvien varchar(255) not null references sinhvien(ma_sinhvien),
    ma_hocphan varchar(255) not null references hocphan(ma_hocphan),
    created_at timestamp default current_timestamp
''')
db.close()
```
