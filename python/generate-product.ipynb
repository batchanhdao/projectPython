```python
import json
import random

def generate_product_names(data):
    with open('product_names.txt', 'a', encoding='utf-8') as file:
        file.writelines(data)
            
with open('product_names.txt', 'r', encoding='utf-8') as file:
    data = file.readlines()

# Tạo tên sản phẩm ngẫu nhiên
product_names_set = set()
remove_text = ['1', '2', '3', '4', '5', '6', '7', '8', '9', '0', '(', ')', '[', ']', '{', '}', '<', '>', '.', ',', ':', ';', '?', '!', '@', '#', '$', '%', '^', '&', '*', '-', '_', '+', '=', '|', '~', '`', '"', "'"]
for d in data:
    text = d.split()
    for t in text:
        check = True
        for r in remove_text:
            if r in t:
                check = False
                break
        if check:
            product_names_set.add(t) 


product_names_list = list(product_names_set)
product_names = []
for i in range(0, 50000):
    name = product_names_list[random.randint(0, len(product_names_list) - 1)] + ' ' + product_names_list[random.randint(0, len(product_names_list) - 1)] + '\n'
    name = str(name).capitalize()
    product_names.append(name)
# print(product_names)
generate_product_names(product_names)


```

```python
from flask import Flask, request, jsonify
from underthesea import word_tokenize, text_normalize, pos_tag
import re
import json
from fuzzywuzzy import process

# app = Flask(__name__)



vietnamese_words_dict = [{'Ngũ cốc': ['lúa mì', 'lúa mạch đen', 'yến mạch', 'ngô', 'bắp', 'lúa mạch', 'kiều mạch', 'gạo', 'ngũ cốc', 'bỏng ngô', 'mì ống', 'macaroni', 'mì sợi', 'mì spaghetti', 'bún', 'bánh ravioli', 'bánh bao', 'bột mì', 'bột nhào', 'bột đánh', 'hỗn hợp làm bánh', 'bột mì đen', 'bột mì nguyên cám', 'bột mì lúa mạch', 'kê', 'bo bo', 'xôi', 'xôi gấc', 'xôi đậu xanh', 'xôi đậu đỏ', 'xôi đậu đen', 'xôi đậu nành', 'xôi đậu phộng', 'xôi ruốc', 'xôi lá cẩm', 'xôi lá sen', 'xôi lá dứa', 'xôi lá chuối', 'xôi 3 màu', 'xôi 4 màu', 'xôi 5 màu', 'xôi 6 màu', 'xôi 7 màu', 'xôi 8 màu', 'xôi 9 màu', 'xôi 10 màu', 'xôi ngọt']}, {'Bánh mỳ': ['bánh mì', 'bánh mì trắng', 'bánh mì lúa mì nguyên cám', 'bánh mì lúa mạch đen', 'bánh mì nho khô', 'bánh mì tỏi', 'bánh mì ngô', 'bánh mì chua', 'bánh mì pháp', 'bánh mì ổ pháp', 'bánh mì pita', 'bánh ngô', 'bánh mì sandwich', 'bánh mì sandwich thịt', 'bánh mì sandwich thịt bò', 'bánh mì sandwich thịt lợn', 'bánh mì sandwich thịt gà', 'bánh mì sandwich thịt heo', 'bánh mì sandwich thịt cừu', 'bánh mỳ kẹp thịt', 'bánh mỳ kẹp thịt bò', 'bánh mỳ kẹp thịt lợn', 'bánh mỳ kẹp thịt gà', 'bánh mỳ kẹp thịt heo', 'bánh mỳ kẹp thịt cừu', 'bánh mỳ kẹp thịt bò nướng', 'bánh mỳ kẹp thịt lợn nướng', 'bánh mỳ kẹp thịt gà nướng', 'bánh mỳ kẹp thịt heo nướng', 'bánh mỳ kẹp thịt cừu nướng']}, {'Bánh cuộn, bánh bao': ['bánh cuộn', 'bánh mì cuộn', 'bánh cuộn vừng', 'bánh cuộn hạt anh túc', 'bánh cuộn quế', 'bánh sừng bò', 'bánh hamburger', 'bánh hot dog', 'bánh tráng cuốn', 'bánh bao', 'bánh bao nhân thịt', 'bánh bao nhân trứng', 'bánh bao nhân đậu', 'bánh bao chay', 'bánh bao nhân hải sản', 'bánh bao nhân thịt bò', 'bánh bao nhân thịt lợn', 'bánh bao nhân thịt gà', 'bánh trung thu', 'bánh trung thu nhân thịt', 'bánh trung thu nhân đậu', 'bánh trung thu nhân hạt sen', 'bánh trung thu nhân hạt dẻ', 'bánh trung thu nhân vừng', 'bánh trung thu nhân đậu xanh', 'bánh trung thu nhân thịt bò', 'bánh trung thu nhân thịt lợn', 'bánh trung thu nhân thịt gà', 'bánh trung thu nhân hải sản', 'bánh trung thu nhân đậu đen', 'bánh trung thu nhân đậu nành', 'bánh trung thu nhân đậu phộng', 'bánh trung thu nhân trứng', 'bánh trung thu dẻo', 'bánh trung thu nướng', 'bánh chưng', 'bánh chưng nhân thịt', 'bánh chưng nhân đậu', 'bánh chưng nhân hạt sen', 'bánh chưng nhân hạt dẻ', 'bánh chưng nhân vừng', 'bánh chưng nhân đậu xanh', 'bánh chưng nhân thịt bò', 'bánh chưng nhân thịt lợn', 'bánh chưng nhân thịt gà', 'bánh chưng nhân hải sản', 'bánh chưng nhân đậu đen', 'bánh chưng nhân đậu nành', 'bánh chưng nhân đậu phộng', 'bánh chưng nhân trứng', 'bánh tét', 'bánh tét nhân thịt', 'bánh tét nhân đậu', 'bánh tét nhân hạt sen', 'bánh tét nhân hạt dẻ', 'bánh tét nhân vừng', 'bánh tét nhân đậu xanh', 'bánh tét nhân thịt bò', 'bánh tét nhân thịt lợn', 'bánh tét nhân thịt gà', 'bánh tét nhân hải sản', 'bánh tét nhân đậu đen', 'bánh tét nhân đậu nành', 'bánh tét nhân đậu phộng', 'bánh tét nhân trứng', 'bánh tét ngọt', 'bánh giầy', 'bánh giầy nhân thịt', 'bánh giầy nhân đậu', 'bánh giầy nhân hạt sen', 'bánh giầy nhân hạt dẻ', 'bánh giầy nhân vừng', 'bánh giầy nhân đậu xanh', 'bánh giầy nhân thịt bò', 'bánh giầy nhân thịt lợn', 'bánh giầy nhân thịt gà', 'bánh giầy nhân hải sản', 'bánh giầy nhân đậu đen', 'bánh giầy nhân đậu nành', 'bánh giầy nhân đậu phộng', 'bánh giầy nhân trứng', 'bánh giầy ngọt', 'bánh giầy nhân chả lụa', 'bánh giầy nhân chả cá', 'bánh giầy nhân giò lụa', 'bánh giò', 'bánh tiêu', 'bánh tiêu nhân thịt', 'bánh tiêu nhân đậu', 'bánh tiêu nhân hạt sen', 'bánh tiêu nhân hạt dẻ', 'bánh tiêu nhân vừng', 'bánh tiêu nhân đậu xanh', 'bánh tiêu nhân thịt bò', 'bánh tiêu nhân thịt lợn', 'bánh tiêu nhân thịt gà', 'bánh tiêu nhân hải sản', 'bánh tiêu nhân đậu đen', 'bánh tiêu nhân đậu nành', 'bánh tiêu nhân đậu phộng', 'bánh tiêu nhân trứng', 'bánh tiêu ngọt', 'bánh tai heo', 'bánh da lợn', 'bánh bông lan', 'bánh bông lan nhân thịt', 'bánh bông lan nhân đậu', 'bánh bông lan nhân hạt sen', 'bánh bông lan nhân hạt dẻ', 'bánh bông lan nhân vừng', 'bánh bông lan nhân đậu xanh', 'bánh bông lan nhân thịt bò', 'bánh bông lan nhân thịt lợn', 'bánh bông lan nhân thịt gà', 'bánh bông lan nhân hải sản', 'bánh bông lan nhân đậu đen', 'bánh bông lan nhân đậu nành', 'bánh bông lan nhân đậu phộng', 'bánh bông lan nhân trứng', 'bánh bông lan ngọt', 'bánh bông lan nhân sô cô la', 'bánh đúc', 'bánh đúc nhân thịt', 'bánh đúc nhân đậu', 'bánh đúc nhân hạt sen', 'bánh đúc nhân hạt dẻ', 'bánh đúc nhân vừng', 'bánh đúc nhân đậu xanh', 'bánh đúc nhân thịt bò', 'bánh đúc nhân thịt lợn', 'bánh đúc nhân thịt gà', 'bánh đúc nhân hải sản', 'bánh đúc nhân đậu đen', 'bánh đúc nhân đậu nành', 'bánh đúc nhân đậu phộng', 'bánh đúc nhân trứng', 'bánh bột lọc', 'bánh bột lọc nhân thịt', 'bánh bột lọc nhân đậu', 'bánh bột lọc nhân hạt sen', 'bánh bột lọc nhân hạt dẻ', 'bánh bột lọc nhân vừng', 'bánh bột lọc nhân đậu xanh', 'bánh bột lọc nhân thịt bò', 'bánh bột lọc nhân thịt lợn', 'bánh bột lọc nhân thịt gà', 'bánh bột lọc nhân hải sản', 'bánh bột lọc nhân đậu đen', 'bánh bột lọc nhân đậu nành', 'bánh bột lọc nhân đậu phộng', 'bánh bột lọc nhân trứng', 'bánh tẻ', 'bánh tẻ nhân thịt', 'bánh tẻ nhân đậu', 'bánh tẻ nhân hạt sen', 'bánh tẻ nhân hạt dẻ', 'bánh tẻ nhân vừng', 'bánh tẻ nhân đậu xanh', 'bánh tẻ nhân thịt bò', 'bánh tẻ nhân thịt lợn', 'bánh tẻ nhân thịt gà', 'bánh tẻ nhân hải sản', 'bánh tẻ nhân đậu đen', 'bánh tẻ nhân đậu nành', 'bánh tẻ nhân đậu phộng', 'bánh tẻ nhân trứng', 'bánh tẻ ngọt', 'bánh tổ', 'bánh tổ nhân thịt', 'bánh tổ nhân đậu', 'bánh tổ nhân hạt sen', 'bánh tổ nhân hạt dẻ', 'bánh tổ nhân vừng', 'bánh tổ nhân đậu xanh', 'bánh tổ nhân thịt bò', 'bánh tổ nhân thịt lợn', 'bánh tổ nhân thịt gà', 'bánh tổ nhân hải sản', 'bánh tổ nhân đậu đen', 'bánh tổ nhân đậu nành', 'bánh tổ nhân đậu phộng', 'bánh tổ nhân trứng', 'bánh tổ ngọt', 'bánh bò', 'bánh bò nhân thịt', 'bánh bò nhân đậu', 'bánh bò nhân hạt sen', 'bánh bò nhân hạt dẻ', 'bánh bò nhân vừng', 'bánh bò nhân đậu xanh', 'bánh bò nhân thịt bò', 'bánh bò nhân thịt lợn', 'bánh bò nhân thịt gà', 'bánh bò nhân hải sản', 'bánh bò nhân đậu đen', 'bánh bò nhân đậu nành', 'bánh bò nhân đậu phộng', 'bánh bò nhân trứng', 'bánh bò ngọt', 'bánh bèo', 'bánh bèo nhân thịt', 'bánh bèo nhân đậu', 'bánh bèo nhân hạt sen', 'bánh bèo nhân hạt dẻ', 'bánh bèo nhân vừng', 'bánh bèo nhân đậu xanh', 'bánh bèo nhân thịt bò', 'bánh bèo nhân thịt lợn', 'bánh bèo nhân thịt gà', 'bánh bèo nhân hải sản', 'bánh bèo nhân đậu đen', 'bánh bèo nhân đậu nành', 'bánh bèo nhân đậu phộng', 'bánh bèo nhân trứng', 'bánh bèo ngọt', 'bánh lá gai', 'bánh lá gai nhân thịt', 'bánh lá gai nhân đậu', 'bánh lá gai nhân hạt sen', 'bánh lá gai nhân hạt dẻ', 'bánh lá gai nhân vừng', 'bánh lá gai nhân đậu xanh', 'bánh lá gai nhân thịt bò', 'bánh lá gai nhân thịt lợn', 'bánh lá gai nhân thịt gà', 'bánh lá gai nhân hải sản', 'bánh lá gai nhân đậu đen', 'bánh lá gai nhân đậu nành', 'bánh lá gai nhân đậu phộng', 'bánh lá gai nhân trứng', 'bánh lá gai ngọt', 'bánh ít trần', 'bánh ít trần nhân thịt', 'bánh ít trần nhân đậu', 'bánh ít trần nhân hạt sen', 'bánh ít trần nhân hạt dẻ', 'bánh ít trần nhân vừng', 'bánh ít trần nhân đậu xanh', 'bánh ít trần nhân thịt bò', 'bánh ít trần nhân thịt lợn', 'bánh ít trần nhân thịt gà', 'bánh ít trần nhân hải sản', 'bánh ít trần nhân đậu đen', 'bánh ít trần nhân đậu nành', 'bánh ít trần nhân đậu phộng', 'bánh ít trần nhân trứng', 'bánh ít trần ngọt', 'bánh trôi', 'bánh trôi nhân đậu xanh', 'bánh trôi nhân đậu đen', 'bánh trôi nhân đậu nành', 'bánh trôi nhân đậu phộng', 'bánh trôi nhân hạt sen', 'bánh trôi nhân hạt dẻ', 'bánh trôi mè đen', 'bánh trôi mè trắng', 'bánh trôi dừa', 'bánh cuốn', 'bánh cuốn nhân thịt', 'bánh cuốn nhân đậu', 'bánh cuốn nhân hạt sen', 'bánh cuốn nhân hạt dẻ', 'bánh cuốn nhân vừng', 'bánh cuốn nhân đậu xanh', 'bánh cuốn nhân thịt bò', 'bánh cuốn nhân thịt lợn', 'bánh cuốn nhân thịt gà', 'bánh cuốn nhân hải sản', 'bánh cuốn nhân đậu đen', 'bánh cuốn nhân đậu nành', 'bánh cuốn nhân đậu phộng', 'bánh cuốn nhân trứng', 'bánh cuốn ngọt', 'bánh cuốn nhân sô cô la', 'bánh cuốn nhân dừa', 'bánh xèo', 'bánh xèo nhân thịt', 'bánh xèo nhân đậu', 'bánh xèo nhân hạt sen', 'bánh xèo nhân hạt dẻ', 'bánh xèo nhân vừng', 'bánh xèo nhân đậu xanh', 'bánh xèo nhân thịt bò', 'bánh xèo nhân thịt lợn', 'bánh xèo nhân thịt gà', 'bánh xèo nhân hải sản', 'bánh xèo nhân đậu đen', 'bánh xèo nhân đậu nành', 'bánh xèo nhân đậu phộng', 'bánh xèo nhân trứng', 'bánh xèo ngọt', 'bánh cốm', 'bánh cốm nhân đậu', 'bánh cốm nhân hạt sen', 'bánh cốm nhân hạt dẻ', 'bánh cốm nhân vừng', 'bánh cốm nhân đậu xanh', 'bánh cốm nhân đậu đen', 'bánh cốm nhân đậu nành', 'bánh cốm nhân đậu phộng', 'bánh xu xê', 'bánh xu xê nhân đậu', 'bánh xu xê nhân hạt sen', 'bánh xu xê nhân hạt dẻ', 'bánh xu xê nhân vừng', 'bánh xu xê nhân đậu xanh', 'bánh xu xê nhân đậu đen', 'bánh xu xê nhân đậu nành', 'bánh xu xê nhân đậu phộng', 'bánh khúc', 'bánh khúc nhân đậu', 'bánh khúc nhân hạt sen', 'bánh khúc nhân hạt dẻ', 'bánh khúc nhân vừng', 'bánh khúc nhân đậu xanh', 'bánh khúc nhân đậu đen', 'bánh khúc nhân đậu nành', 'bánh khúc nhân đậu phộng', 'bánh khúc nhân thịt', 'bánh khúc nhân thịt bò', 'bánh khúc nhân thịt lợn', 'bánh khúc nhân thịt gà', 'bánh khúc nhân hải sản', 'bánh khúc nhân trứng', 'bánh khúc ngọt', 'bánh chè lam', 'bánh chè lam nhân đậu', 'bánh chè lam nhân hạt sen', 'bánh chè lam nhân hạt dẻ', 'bánh chè lam nhân vừng', 'bánh chè lam nhân đậu xanh', 'bánh chè lam nhân đậu đen', 'bánh chè lam nhân đậu nành', 'bánh chè lam nhân đậu phộng', 'bánh chuối', 'bánh táo', 'bánh hạt dẻ', 'bánh đậu xanh', 'bánh đậu đen', 'bánh đậu nành', 'bánh phô mai', 'bánh hạt sen', 'bánh vừng', 'bánh dừa', 'bánh hạt điều', 'bánh hạt lạc']}, {'Bánh ngọt và bánh quy': ['bánh nướng', 'bánh táo', 'bánh việt quất', 'bánh anh đào', 'bánh tự làm', 'bánh tart', 'bánh tart táo', 'bánh nhân thịt bằm', 'bánh bí ngô', 'bánh nhân đại hoàng', 'bánh nhân thịt', 'bánh knish', 'bánh pizza', 'bánh muffin', 'bánh muffin việt quất', 'bánh muffin nho khô', 'bánh quy', 'bánh quy kem chua', 'bánh kếp', 'bánh muffin anh', 'bánh rán', 'bánh quế', 'bánh kem', 'bánh sinh nhật', 'bánh cưới', 'bánh giáng sinh', 'bánh trái cây', 'bánh kem dâu tây', 'bánh sô cô la', 'bánh sô cô la chip', 'bánh sô cô la nhiều lớp', 'bánh phủ sô cô la', 'bánh mật ong', 'bánh pound', 'bánh hạnh nhân', 'bánh napoleon', 'bánh bông lan', 'bánh phô mai', 'bánh chanh', 'bánh ngọt', 'bánh gừng', 'bánh cà phê', 'bánh nướng nhỏ nho', 'bánh brownie fudge', 'bánh quy yến mạch', 'bánh quy sô cô la']}, {'thịt gia súc': ['thịt', 'thịt bò', 'thịt lợn', 'thịt bê', 'thịt cừu', 'thịt cừu non', 'thịt bò bít tết', 'thịt bò nướng', 'thịt bò xay', 'sườn heo', 'sườn cừu', 'cốt lết bê', 'giăm bông', 'thịt xông khói', 'thịt hun khói pastrami', 'thịt bò muối', 'xúc xích', 'salami', 'xúc xích hun khói', 'bologna', 'xúc xích liên kết', 'xúc xích đức', 'xúc xích wiener']}, {'Gia cầm': ['gia cầm', 'thịt gà', 'thịt gà ri', 'thịt gà đông tảo', 'thịt gà hồ', 'thịt gà mía', 'thịt gà tàu vàng', 'thịt gà ác', 'thịt gà tre', 'thịt gà tây', 'thịt ngỗng', 'thịt vịt', 'thịt vịt cỏ', 'thịt vịt bầu', 'trứng gà', 'trứng gà ri', 'trứng gà đông tảo', 'trứng gà hồ', 'trứng gà mía', 'trứng gà tàu vàng', 'trứng gà ác', 'trứng gà tre', 'trứng gà tây', 'trứng ngỗng', 'trứng vịt', 'trứng vịt cỏ', 'trứng vịt bầu', 'gà', 'gà tây', 'ngỗng', 'vịt', 'trứng', 'gà nguyên con', 'gà xé nhỏ', 'đùi gà', 'cánh gà', 'ức gà', 'ức gà tây']}, {'Cá và các sản phẩm từ cá': ['cá', 'cá hồi', 'cá hồi vân', 'cá tầm', 'cá tuyết', 'cá bơn', 'cá ngừ', 'cá rô', 'cá vược', 'cá mú', 'cá chó', 'cá trích', 'cá chình', 'cá thu', 'cá cơm', 'cá đối', 'cá chép', 'cá mòi', 'cá da trơn', 'trứng cá muối', 'bít tết cá', 'bít tết cá hồi', 'phi lê cá', 'phi lê cá bơn', 'cá hun khói', 'cá muối', 'cá trích ướp', 'hải sản', 'tôm', 'tôm càng', 'cua', 'tôm hùm', 'hàu', 'nghêu', 'động vật có vỏ', 'mực', 'nước mắm', 'mắm tôm', 'mắm cá', 'mắm tép', 'mắm ruốc', 'mắm nêm']}, {'Sữa': ['sữa', 'sữa nguyên chất', 'sữa tách kem', 'sữa ít béo', 'sữa không béo', 'sữa tiệt trùng', 'sữa bò', 'sữa đậu nành', 'sữa dừa', 'sữa hạt', 'sữa hạt sen', 'sữa hạt dẻ', 'sữa hạt điều', 'sữa hạt lanh', 'sữa hạt mắc ca', 'sữa hạt óc chó', 'sữa đậu đen', 'sữa bột', 'sữa đặc', 'sữa chua', 'kefir', 'bơ sữa', 'kem', 'kem chua', 'bơ', 'phô mai tươi', 'phô mai nông trại', 'phô mai tự làm', 'phô mai kem', 'phô mai thụy sĩ', 'parmesan', 'cheddar', 'mozzarella', 'roquefort', 'phô mai xanh', 'phô mai cứng', 'phô mai mềm', 'phô mai sắc', 'phô mai nhẹ', 'phô mai hun khói', 'phô mai bào', 'kem vani', 'kem sô cô la', 'kem trái cây', 'kem xoài', 'kem táo', 'kem nho', 'kem dừa', 'kem đu đủ', 'kem dưa hấu', 'kem bưởi', 'kem chuối', 'kem kiwi', 'kem vải', 'kem đào', 'kem chanh', 'kem cam', 'kem dâu tây', 'kem ốc quế', 'kem que', 'kem sundae', 'đá bào', 'kem bơ', 'đá bào 5 màu', 'đá bào 6 màu', 'đá bào 7 màu', 'đá bào 8 màu', 'đá bào 9 màu']}, {'Hoa quả': ['trái cây tươi', 'táo', 'lê', 'mơ', 'đào', 'xuân đào', 'mận', 'nho', 'anh đào', 'anh đào ngọt', 'chanh', 'chanh xanh', 'cam', 'cam sành', 'cam vinh', 'cam hà tĩnh', 'cam mật', 'cam canh', 'cam cao phong', 'quýt', 'bưởi', 'bưởi năm roi', 'bưởi da xanh', 'bưởi diễn', 'bưởi tân triều', 'bưởi luận văn', 'chuối', 'chuối chín', 'chuối xanh', 'chuối hột', 'chuối tiêu', 'chuối sứ', 'chuối ngự', 'kiwi', 'dứa', 'ô liu', 'sung', 'đu đủ', 'xoài', 'xoài tứu quý', 'xoài bao tử', 'xoài thái', 'xoài cát', 'xoài tượng', 'xoài úc', 'bơ', 'dừa', 'lựu', 'dưa', 'dưa hấu', 'dưa lưới', 'dưa leo', 'dưa hồng', 'dưa gang', 'dưa chuột', 'nho', 'nho ninh thuận', 'nho móng tay', 'nho thân gỗ', 'nho pháp', 'nho rừng', 'nho mỹ', 'vải', 'nhãn', 'mâm xôi', 'mâm xôi đỏ', 'quả lê', 'quả trứng gà', 'quả sung', 'quả dừa', 'quả hồng giòn', 'quả hồng đỏ', 'quả ổi', 'quả lê']}, {'Quả mọng': ['quả mọng', 'dâu tây', 'quả mâm xôi', 'quả nam việt quất', 'quả việt quất', 'quả việt quất đen', 'nho đen', 'nho đỏ', 'lý gai', 'mâm xôi đen']}, {'Trái cây sấy khô': ['trái cây sấy khô', 'mơ khô', 'nho khô', 'sung', 'mận khô', 'chà là', 'trái cây kẹo', 'mít sấy', 'chuói sấy', 'dừa sấy', 'xoài sấy', 'mâm xôi sấy', 'mâm xôi đỏ sấy', 'nho khô', 'táo sấy', 'việt quất sấy', 'mận sấy', 'chà là sấy', 'khoai lang sấy', 'khoai tây sấy', 'cà chua sấy', 'cà rốt sấy', 'cà tím sấy']}, {'Các loại hạt và hạt giống': ['các loại hạt', 'hạt phỉ', 'quả óc chó', 'hạnh nhân', 'hạt dẻ', 'đậu phộng', 'hạt dẻ cười', 'hạt tuyết tùng', 'hạt điều', 'hạt hồ đào', 'hạt macadamia', 'hạt mơ', 'hạt bí ngô', 'hạt hướng dương', 'hạt lanh', 'hạt mắc ca', 'hạt óc chó', 'hạt óc chó đỏ', 'hạt mè', 'hạt mè đen', 'hạt mè trắng', 'hạt lạc', 'hạt cải', 'hạt đậu đen', 'hạt đậu xanh', 'hạt đậu nành']}, {'Bảo quản': ['mứt mâm xôi', 'mứt nam việt quất', 'thạch nho', 'mứt cam', 'mứt nho', 'mứt dâu', 'mứt mơ', 'mứt xoài', 'mứt khoai lang', 'mật ong', 'xi-rô cây phong', 'bơ đậu phộng']}, {'Rau': ['rau tươi', 'rau trộn', 'rau đóng hộp', 'rau lá', 'rau xanh', 'cà chua', 'dưa chuột', 'cà rốt', 'củ cải đường', 'khoai tây', 'hành tây', 'hành lá', 'tỏi tây', 'ớt ngọt', 'ớt đỏ', 'ớt xanh', 'ớt vàng', 'ớt bột', 'ớt cay', 'bắp cải', 'súp lơ', 'bông cải xanh', 'cải brussels', 'cải xanh', 'cải xoăn', 'su hào', 'nấm', 'rau diếp', 'rau bina', 'cần tây', 'măng tây', 'atisô', 'cải xoong', 'rau diếp xoăn', 'rau diếp xoăn', 'tỏi', 'cà tím', 'bí', 'bầu', 'bí xanh', 'bí ngô', 'củ cải', 'củ cải đường', 'củ cải đường', 'cải ngựa', 'dưa chuột muối', 'dưa chua', 'dưa chuột ướp', 'dưa cải bắp', 'ô liu đóng hộp', 'rau ngót', 'rau dền', 'rau muống', 'rau mùi tàu', 'rau răm', 'rau mùi', 'rau lang', 'rau cải']}, {'Đậu Hà Lan, đậu, ngô': ['đậu xanh', 'đậu ngọt', 'đậu que', 'đậu xanh', 'đậu lima', 'đậu thận', 'đậu đen', 'đậu nành', 'đậu lăng', 'ngô', 'ngô ngọt', 'ngô', 'hạt cà phê']}, {'Thảo mộc': ['thì là', 'rau mùi tây', 'húng quế', 'rau mùi', 'bạc hà', 'hẹ', 'mã đề', 'tía tô', 'lá mơ lông', 'lá sung', 'lá ổi']}, {'Nước ép trái cây, đồ uống': ['nước ép táo', 'nước ép cam', 'nước ép bưởi', 'nước ép chanh', 'nước ép cà chua', 'nước ép cà rốt', 'nước ép cà tím', 'nước ép dưa hấu', 'nước ép lựu', 'nước ép mâm xôi', 'nước ép mơ', 'nước ép nho', 'nước ép ổi', 'nước ép quýt', 'nước ép đu đủ', 'nước ép xoài', 'nước ép việt quất', 'nước ép việt quất đen', 'nước ép việt quất đỏ', 'nước ép trái cây tươi', 'trà', 'trà xanh', 'trà đen', 'trà sữa', 'trà đá', 'trà thảo mộc', 'trà bạc hà', 'trà ấn độ', 'cà phê', 'cà phê hòa tan', 'cà phê espresso', 'cà phê cappuccino', 'cà phê không chứa caffein', 'cà phê đen', 'cà phê sữa', 'cà phê kem', 'nửa sữa nửa kem', 'ca cao', 'sô cô la nóng', 'sữa lắc']}, {'Nước, nước ngọt': ['nước khoáng', 'nước suối', 'nước ngọt', 'nước soda', 'nước chanh', 'rượu táo', 'bia gừng', 'nước mía', 'nước dừa']}, {'Đồ uống có cồn': ['bia', 'bia đen', 'rượu vang', 'rượu vang đỏ', 'rượu vang trắng', 'rượu sâm panh', 'rượu vodka', 'rượu cognac', 'rượu mạnh', 'rượu whisky', 'rượu gin', 'rượu rum', 'rượu mùi', 'cocktail', 'rượu đấm', 'rượu nho', 'rượu hồng', 'rượu nếp', 'rượu đế', 'rượu rắn', 'rượu rắn hổ mang', 'rượu mẫu sơn', 'rượu làng vân', 'rượu bầu', 'rượu vang đà lạt']}, {'Nước sốt, nước trộn salad, dầu thực vật, chất béo': ['sốt cà chua', 'sốt cà chua', 'sốt nấm', 'sốt thịt', 'sốt bít tết', 'nước sốt thịt', 'sốt mì spaghetti', 'nước sốt cay', 'nước sốt ớt', 'nước sốt barbecue', 'nước sốt chua ngọt', 'nước sốt tỏi', 'nước sốt trắng', 'nước chấm', 'nước tương', 'nước sốt táo', 'nước sốt nam việt quất', 'sốt mayonnaise', 'nước sốt trộn salad', 'nước sốt nga', 'nước sốt ý', 'nước sốt pháp', 'nước sốt phô mai xanh', 'dầu thực vật', 'dầu ô liu', 'dầu ngô', 'dầu hạt hướng dương', 'dầu mè', 'bơ thực vật', 'mỡ', 'mỡ lợn', 'chất béo', 'mỡ động vật', 'mỡ thực vật', 'chin su', 'sa tế']}, {'Gia vị và gia vị': ['gia vị', 'hương liệu', 'gia vị', 'tương ớt', 'thảo mộc', 'hạt', 'giấm', 'hạt tiêu', 'hạt tiêu xay', 'hạt tiêu nguyên hạt', 'ớt đỏ', 'ớt cay', 'muối', 'thì là', 'rau mùi tây', 'bạc hà', 'rau mùi', 'húng quế', 'lá nguyệt quế', 'đinh hương', 'hạt nhục đậu khấu', 'quế', 'hạt caraway', 'húng tây', 'bạch đậu khấu', 'ngải giấm', 'kinh giới cay', 'hương thảo', 'tỏi', 'mù tạt', 'vỏ chanh']}, {'Kẹo, sô cô la': ['kẹo', 'kẹo', 'đồ ngọt', 'kẹo caramel', 'kẹo bạc hà', 'kẹo dẻo', 'kẹo que', 'kẹo bonbon', 'sôcôla', 'kẹo sôcôla', 'sôcôla', 'thanh sôcôla', 'thanh kẹo', 'kẹo bơ cứng', 'kẹo mềm', 'kẹo dẻo', 'kẹo cu đơ', 'kẹo gạo lứt', 'kẹo mè xửng', 'kẹo dừa', 'bánh pía sóc trăng', 'kẹo lạc']}, {'Quần áo': ['quần áo nữ', 'quần áo nam', 'quần áo trẻ em', 'quần áo trẻ sơ sinh', 'quần áo thể thao', 'quần áo mùa đông', 'quần áo mùa hè', 'quần áo thường ngày', 'quần áo trang trọng', 'quần áo đắt tiền', 'quần áo rẻ tiền', 'quần áo đẹp', 'quần áo tồi tàn', 'quần áo thiết kế', 'quần áo may đo', 'quần áo thủ công', 'quần áo may sẵn', 'quần áo bảo hộ lao động', 'đồng phục', 'giày dép', 'mũ', 'đồ lót', 'đồ thể thao', 'đồ bơi', 'đồ ngủ', 'thời trang', 'phong cách', 'mốt', 'thời trang cao cấp']}, {'Áo khoác': ['áo khoác mùa đông', 'áo khoác ấm', 'áo khoác lông', 'áo khoác da cừu', 'áo khoác chống bão', 'áo khoác parka', 'áo khoác tuyết', 'áo khoác dài', 'áo mưa', 'áo khoác chống thấm nước', 'áo khoác da', 'áo khoác bomber', 'áo khoác xe hơi', 'áo gió', 'áo khoác thể thao']}, {'Quần áo nữ': ['trang phục thường ngày', 'trang phục trang trọng', 'trang phục mùa hè', 'trang phục len', 'váy cocktail', 'váy khoét sâu', 'váy dạ hội', 'bộ đồ công sở', 'bộ đồ vest', 'đồng phục', 'váy cài nút', 'váy yếm', 'váy mặc ở nhà', 'đồ bầu', 'váy bầu', 'áo len', 'áo len dài tay', 'áo len cardigan', 'áo khoác', 'áo vest', 'áo cánh', 'áo sơ mi', 'áo dài', 'áo phông', 'áo ba lỗ', 'váy', 'váy ngắn', 'váy dài', 'váy xếp ly', 'váy đen', 'quần culottes', 'quần dài', 'quần jean', 'quần denim', 'quần short', 'đồ bơi', 'bikini', 'bộ đồ liền thân', 'bộ đồ khởi động', 'bộ đồ tập luyện']}, {'Quần áo nam': ['bộ đồ công sở', 'bộ đồ ba mảnh', 'áo tuxedo', 'cà vạt đen', 'cà vạt trắng', 'quần dài', 'quần jean', 'quần denim', 'quần short', 'áo khoác', 'áo khoác blazer', 'áo vest', 'áo len', 'áo len cổ lọ', 'áo sơ mi', 'áo polo', 'áo phông', 'áo ba lỗ', 'đồ tập luyện', 'quần tắm', 'quần bơi', 'đồng phục', 'quần áo làm việc', 'quần yếm']}, {'Đồ lót và áo choàng': ['đồ lót', 'quần lót', 'áo ngực', 'váy lót', 'nửa váy lót', 'quần lót', 'áo lót', 'quần tất', 'quần bó', 'quần legging', 'tất', 'tất dài', 'tất cao đến đầu gối', 'đồ ngủ', 'áo ngủ', 'áo choàng', 'áo choàng tắm', 'áo choàng vải bông']}, {'Phụ kiện': ['túi xách', 'túi xách tay', 'ví', 'ví cầm tay', 'cặp', 'cặp tài liệu', 'ví đựng tiền', 'ví đựng tiền xu', 'bộ đồ trang điểm', 'hộp đựng mỹ phẩm', 'bộ dao cạo râu', 'bộ đồ du lịch', 'túi du lịch', 'túi vải thô', 'túi thể thao', 'ba lô', 'ba lô đeo vai', 'túi tote', 'hành lý xách tay', 'túi xách mang theo', 'vali', 'túi đựng quần áo', 'rương đựng đồ', 'thắt lưng', 'dây đeo', 'ô', 'nút', 'dây kéo', 'khóa', 'móc cài', 'miếng dán velcro', 'mũ', 'mũ lông', 'mũ nồi', 'mũ lưỡi trai', 'mũ bóng chày', 'mũ trùm đầu', 'băng đô', 'mũ bảo hiểm', 'khăn quàng cổ', 'khăn tay', 'khăn lụa', 'khăn len', 'khăn đan', 'găng tay', 'bao tay', 'bịt tai', 'kính râm', 'kính mắt', 'kính áp tròng', 'cà vạt', 'cà vạt lụa', 'nơ']}, {'Giày dép': ['giày bốt mùa đông', 'giày bốt nữ', 'giày bốt đi bộ đường dài', 'giày bốt cao su', 'giày đi bộ', 'giày oxford', 'giày lười', 'giày moccasin', 'giày cao gót', 'giày dạ hội', 'giày da bóng', 'giày da lộn', 'giày gót thấp', 'giày gót bằng', 'dép xăng đan', 'giày hở mũi', 'dép lê', 'giày đế xuồng', 'dép cao su', 'giày tennis', 'giày chạy bộ', 'giày thể thao', 'giày tập luyện', 'đế giày', 'đế trong', 'lớp lót', 'đệm', 'gót giày', 'lưỡi giày', 'khóa giày', 'dây giày']}, {'Trang sức': ['nhẫn vàng', 'nhẫn kim cương', 'nhẫn đính hôn', 'nhẫn cưới', 'vòng cổ', 'vòng tay', 'trâm cài', 'hoa tai', 'dây chuyền', 'mặt dây chuyền', 'bùa hộ mệnh', 'khuy măng sét', 'kẹp cà vạt', 'ghim cài cà vạt', 'đồng hồ', 'đồng hồ đeo tay', 'đồ trang sức', 'đá quý', 'đá bán quý', 'kim cương', 'hồng ngọc', 'ngọc bích', 'ngọc lục bảo', 'alexandrite', 'opal', 'topaz', 'ngọc trai']}, {'Thiết bị văn phòng': ['máy tính', 'máy in', 'máy quét', 'máy đánh chữ', 'máy tính bỏ túi', 'máy photocopy', 'máy xerox', 'máy hủy giấy', 'giấy viết', 'giấy ghi chú', 'giấy a4', 'giấy vẽ', 'giấy than', 'giấy có dòng kẻ', 'giấy kẻ ô', 'giấy không có dòng kẻ', 'giấy nháp', 'giấy thấm', 'giấy trong suốt', 'giấy can', 'giấy màu', 'giấy nâu', 'giấy gói', 'sổ tay', 'tập viết', 'sổ ghi chép', 'tập giấy nháp', 'bảng kẹp', 'bảng vẽ', 'hồ sơ', 'thư mục', 'bìa cứng', 'dụng cụ sắp xếp', 'tủ hồ sơ', 'khay đựng thư', 'phong bì', 'bút bi', 'bút máy', 'bút chì', 'bút dạ màu', 'bút sáp màu', 'gọt bút chì', 'cục tẩy', 'thước kẻ', 'giấy nến', 'dao rọc giấy', 'dao bỏ túi', 'vật chặn giấy', 'kẹp giấy', 'máy bấm kim', 'đinh ghim', 'băng dính', 'nhãn dán', 'nhãn', 'giấy vụn', 'thùng đựng rác', 'giỏ đựng giấy vụn']}, {'Đồ gia dụng, đồ điện': ['đèn chiếu sáng', 'đèn sàn', 'đèn bàn', 'đèn đọc sách', 'đèn chùm', 'bóng đèn', 'đèn pin', 'tivi', 'hệ thống âm thanh nổi', 'radio', 'đồng hồ', 'đồng hồ báo thức', 'máy khâu', 'máy điều hòa', 'quạt', 'máy thông gió', 'máy sưởi', 'bàn là', 'máy sấy tóc', 'tủ lạnh', 'tủ đông', 'lò vi sóng', 'máy nướng bánh mì', 'bếp điện', 'lò nướng', 'nồi áp suất', 'máy trộn', 'máy xay sinh tố', 'máy chế biến thực phẩm', 'máy xay thịt', 'máy băm thịt', 'ấm đun nước điện', 'máy pha cà phê', 'máy xay cà phê', 'máy giặt', 'máy rửa chén', 'máy hút bụi', 'máy nước nóng', 'nồi hơ']}, {'Đồ dùng nhà bếp': ['ấm trà', 'ấm pha cà phê', 'bình', 'lọ', 'bình đựng', 'cốc cà phê', 'cốc bia', 'đồ sứ', 'tách trà', 'tách cà phê', 'đĩa lót', 'bát đựng đường', 'pha lê', 'đồ thủy tinh', 'ly', 'ly rượu', 'cốc', 'cốc vại', 'ly bia', 'đồ dùng trên bàn ăn', 'đĩa', 'khay', 'đĩa ăn tối', 'bát đựng súp', 'nĩa', 'thìa', 'dao kéo', 'dao', 'nĩa cắt thịt', 'dao cắt bánh mì', 'dao cắt hoa quả', 'thớt', 'giá để bát đĩa', 'hộp đựng', 'khăn trải bàn', 'khăn ăn', 'khăn giấy ăn', 'khăn ăn lót đĩa', 'lọ muối', 'lọ tiêu', 'chảo', 'nồi nấu chậm', 'nồi hầm', 'nồi', 'lò nướng kiểu hà lan', 'muôi', 'kẹp gắp', 'dụng cụ khui hộp', 'dụng cụ nghiền khoai tây', 'cán lăn', 'máy đánh trứng', 'dụng cụ nạo', 'kẹp hạt dẻ', 'dụng cụ gọt rau củ', 'khăn lau bát đĩa', 'thùng rác']}, {'Đồ nội thất': ['bàn', 'bàn làm việc', 'bàn viết', 'bàn có nắp cuộn', 'bàn máy tính', 'bàn gấp', 'bàn ăn', 'bàn bếp', 'bàn trà', 'bàn cà phê', 'ghế sofa', 'ghế dài', 'ghế bập bênh', 'ghế văn phòng', 'ghế xoay', 'ghế đẩu bếp', 'giường', 'giường queen', 'giường đôi', 'giường đơn', 'giường tầng', 'giường gấp', 'tủ đầu giường', 'bàn trang điểm', 'gương', 'tủ quần áo', 'tủ ngăn kéo', 'tủ trang điểm', 'tủ đựng đồ', 'tủ bếp', 'thảm', 'thảm phòng tắm']}, {'Mỹ phẩm và đồ vệ sinh': ['kem đánh răng', 'bàn chải đánh răng', 'chỉ nha khoa', 'nước súc miệng', 'xà phòng', 'dầu gội', 'dầu xả', 'thuốc nhuộm tóc', 'lược chải tóc', 'gel', 'mousse', 'chất khử mùi', 'nước hoa', 'kem dưỡng da', 'kem dưỡng da tay', 'kem dưỡng da mặt', 'kem dưỡng ẩm', 'kem lạnh', 'dao cạo', 'kem cạo râu', 'mỹ phẩm', 'son môi', 'son bóng', 'phấn má hồng', 'phấn phủ', 'mascara', 'bút kẻ mắt', 'chì kẻ mắt', 'phấn mắt', 'nước tẩy sơn móng tay', 'kéo cắt móng tay', 'dũa móng tay', 'kem dưỡng móng tay', 'sáp vuốt tóc', 'giấy vệ sinh', 'băng cá nhân']}, {'Máy tính': ['laptop', 'máy tính bảng', 'máy tính xách tay', 'máy tính để bàn', 'bộ vi xử lý', 'ram', 'ổ cứng ssd', 'usb', 'sạc', 'bộ chuyển đổi nguồn', 'bàn phím', 'chuột', 'dây cáp', 'dây sạc']}, {'Điện thoại': ['điện thoại', 'smartphone', 'điện thoại di động', 'điện thoại cảm ứng', 'điện thoại phổ thông', 'điện thoại chơi game', 'điện thoại pin khủng', 'điện thoại chụp ảnh', 'điện thoại selfie', 'điện thoại chống nước', 'điện thoại chống sốc', 'điện thoại chống bụi', 'điện thoại chống va đập']}, {'Máy ảnh': ['máy ảnh', 'máy quay video', 'ống kính máy ảnh', 'thẻ nhớ', 'pin máy ảnh', 'túi đựng máy ảnh', 'dây đeo máy ảnh', 'bộ lau máy ảnh', 'bộ vệ sinh máy ảnh', 'bộ chuyển đổi ống kính', 'bộ chuyển đổi pin']}, {'Loa và Tai nghe': ['loa', 'loa di động', 'loa bluetooth', 'tai nghe', 'loa bluetooth', 'sạc dự phòng', 'tai nghe bluetooth', 'tai nghe không dây', 'tai nghe chống ồn', 'tai nghe chống nước', 'tai nghe chống mồ hôi', 'tai nghe chống bụi', 'tai nghe chống va đập', 'tai nghe chống sốc']}, {'Điều hòa không khí và Quạt điện': ['điều hòa không khí', 'điều hòa treo tường', 'điều hòa trần', 'điều hòa cửa sổ', 'điều hòa âm trần', 'điều hòa đứng', 'quạt điện', 'quạt treo tường', 'quạt trần', 'quạt hơi nước', 'quạt hút', 'quạt thông gió', 'quạt đứng', 'quạt bàn', 'quạt sạc', 'quạt cầm tay', 'quạt hơi nước', 'quạt hơi nước mini', 'quạt hơi nước cầm tay', 'quạt hơi nước treo tường', 'quạt hơi nước trần', 'quạt hơi nước thông gió', 'quạt hơi nước đứng']}, {'Thiết bị gia dụng': ['tủ lạnh', 'tủ đông', 'lò nướng']}, {'Máy giặt và Máy rửa chén': ['máy giặt', 'máy rửa chén', 'máy rửa chén cửa trên', 'máy giặt cửa trên', 'máy giặt cửa trước']}, {'Đồng hồ và Thiết bị sức khỏe': ['đồng hồ thông minh', 'máy đo huyết áp', 'máy theo dõi nhịp tim', 'thiết bị đeo tay', 'thiết bị đo nhịp tim', 'thiết bị đo huyết áp', 'thiết bị đo nhiệt độ', 'thiết bị đo đường huyết', 'thiết bị đo nồng độ oxy', 'thiết bị đo nồng độ co2', 'thiết bị đo nồng độ khí', 'thiết bị đo nồng độ bụi', 'thiết bị đo nồng độ chất độc hại', 'thiết bị đo nồng độ chất ô nhiễm']}, {'Thiết bị chơi game': ['máy tính chơi game', 'máy chơi game console', 'vr headset', 'tai nghe chơi game', 'bàn phím chơi game', 'chuột chơi game', 'bàn điều khiển chơi game', 'thiết bị chơi game cầm tay', 'thiết bị chơi game cầm tay không dây', 'thiết bị chơi game cầm tay có dây', 'thiết bị chơi game cầm tay chống sốc', 'thiết bị chơi game cầm tay chống va đập', 'thiết bị chơi game cầm tay chống nước', 'thiết bị chơi game cầm tay chống bụi', 'thiết bị chơi game cầm tay chống mồ hôi', 'thiết bị chơi game cầm tay chống sốc']}, {'Camera và Thiết bị giám sát': ['camera an ninh', 'camera giám sát', 'camera hành trình', 'máy dò kim loại', 'máy quét mã vạch', 'máy quét mã qr', 'máy quét mã vạch cầm tay', 'máy quét mã vạch không dây', 'máy quét mã vạch có dây', 'máy quét mã vạch chống sốc', 'máy quét mã vạch chống va đập']}, {'Thiết bị điện tử': ['bảng điều khiển', 'thiết bị phát sóng', 'máy phun sương', 'máy đếm tiền', 'máy in hóa đơn', 'bộ giải mã tín hiệu']}]


# Danh sách các stop words
STOP_WORDS = [
    "bị", "đó", "rằng", "bởi", "được", "rất", "cả", "dưới", "rồi", "các", "gì", "sau", 
    "cái", "khi", "sẽ", "cần", "không", "so", "càng", "là", "sự", "chỉ", "lại", "tại", 
    "chiếc", "lên", "theo", "cho", "lúc", "thì", "chứ", "mà", "trên", "chưa", "mỗi", 
    "trước", "chuyện", "một", "từ", "có", "này", "từng", "cứ", "nên", "và", "của", "nếu", 
    "về", "cùng", "ngay", "vẫn", "cũng", "nhiều", "vào", "đã", "như", "vậy", "đang", 
    "nhưng", "vì", "đây", "những", "việc", "để", "nơi", "với", "đến", "nữa", "vừa", 
    "đều", "phải", "điều", "qua", "do", "ra", "ở", "va"
]
class ConvertVietnameseToNormalText:
    def slugify(self, input_text, separator=' '):
        input_text = self.to_non_accent_vietnamese(input_text.strip())
        slug = re.sub(r"[^a-z0-9]+", separator, input_text.lower())
        return slug

    def to_non_accent_vietnamese(self, text):
        """
        Convert a Vietnamese string to a non-accented version.
        """
        # Replace accented characters with their ASCII equivalents
        text = re.sub(r"[AÁÀÃẠÂẤẦẪẬĂẮẰẴẶ]", "A", text)
        text = re.sub(r"[aàáạảãâầấậẩẫăằắặẳẵ]", "a", text)
        text = re.sub(r"[EÉÈẼẸÊẾỀỄỆ]", "E", text)
        text = re.sub(r"[eèéẹẻẽêềếệểễ]", "e", text)
        text = re.sub(r"[IÍÌĨỊ]", "I", text)
        text = re.sub(r"[iìíịỉĩ]", "i", text)
        text = re.sub(r"[OÓÒÕỌÔỐỒỖỘƠỚỜỠỢ]", "O", text)
        text = re.sub(r"[oòóọỏõôồốộổỗơờớợởỡ]", "o", text)
        text = re.sub(r"[UÚÙŨỤƯỨỪỮỰ]", "U", text)
        text = re.sub(r"[uùúụủũưừứựửữ]", "u", text)
        text = re.sub(r"[YÝỲỸỴ]", "Y", text)
        text = re.sub(r"[yỳýỵỷỹ]", "y", text)
        text = re.sub(r"[Đ]", "D", text)
        text = re.sub(r"[đ]", "d", text)
        # Remove combining diacritical marks
        text = re.sub(r"[\u0300\u0301\u0303\u0309\u0323]", "", text)  # Huyền, sắc, hỏi, ngã, nặng
        text = re.sub(r"[\u02C6\u0306\u031B]", "", text)  # Â, Ê, Ă, Ơ, Ư
        return text


def fuzzy_convert(word, vietnamese_words_dict):
    max_len = 0
    query = ""
    category = ""
    converter = ConvertVietnameseToNormalText() 
    # Tìm từ gần nhất trong từ điển
    for list_words in vietnamese_words_dict:
        for key, value in list_words.items():
            matches = process.extractBests(word, value)
            for match in matches:
                if converter.slugify(word) in converter.slugify(match[0]) or (match[1] > 70 and (len(match[0]) >= len(word) - 1 and len(match[0]) <= len(word) + 1 )):
                    query = match[0]
                    category = key
                    print("category: ", category)
                    return {
                        "query": query,
                        "category": category
                    }
                if match[1] > 70 and len(match[0]) > max_len:
                    max_len = len(match[0])
                    query = match[0]
                    category = key
    print("category: ", category)
    return {
        "query": query,
        "category": category
    }
    
  
def convert_with_fuzzy(query):
    words = re.findall(r'\b\w+\b', query.lower())
    # Loại bỏ stop words
    filtered_words = [word for word in words if word not in STOP_WORDS]
    print("filtered words: ", filtered_words)
    
    result = fuzzy_convert(" ".join(filtered_words), vietnamese_words_dict).get("query")
    return result

list_input_query = ["quan ao, the - thao cho tre.", "banh chung thu ngay ram", "may tinh xach tay", "banh trưng", "bánh chung", "bàn chai danh rang", "áo ves", "cam sanh", "cá chich", "banh khuc", "bánh giay"]
for query in list_input_query:
    print("query input:", query)
    print("query output:", pos_tag(convert_with_fuzzy(query)))
    print("=====================================")

```
